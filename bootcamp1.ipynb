```python
X=data.drop('target',axis=1)
y=data['target']
```

```python
X
```

```python
y
```

```python
#create scalar
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_Scaled=scalar.fit_transform(X)
X_Scaled
```

```python
#split data randomly to training and testing data in the ration 75% and 25%
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_Scaled,y,test_size=0.2,random_state=42)
```

```python
X_test
```

```python
X_train, X_test, y_train, y_test
```

```python
#shape of data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#create decision tree model
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
```

```python
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
```

```python
# train DecisionTree classifier - criterion can be gini and entropy
#model = DecisionTreeClassifier(criterion='gini',max_depth=3, random_state=42)
#model.fit(X_train, y_train)
```

```python
#test the model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
#classification report
from sklearn.metrics import classification_report, confusion_matrix , ConfusionMatrixDisplay
print ("Classification Report")
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix Visualization
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
# Visualizing the decision tree
plt.figure(figsize=(16, 10))
plot_tree(model, feature_names=X.columns, class_names=['No Heart Disease', 'Heart Disease'],
          filled=True, rounded=True, fontsize=10, max_depth=2)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
#merge y_pred and y and show matched predictions
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
result
```

```python
#show which did not match
result[result['Actual'] != result['Predicted']]
```

```python
#find accuracy
from sklearn.metrics import accuracy_score
```

```python
#compare y_test and y_pred
accuracy = accuracy_score(y_test, y_pred)
```
