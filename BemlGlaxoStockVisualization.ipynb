```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#import data
beml_df = pd.read_csv('/content/sample_data/BEML.csv')
glaxo_df = pd.read_csv('/content/sample_data/GLAXO.csv')
```

```python
beml_df.head()
```

```python
glaxo_df.head()
```

```python
#filter
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]

```

```python
beml_df.dtypes
```

```python
#Create a new column with date as index - set date time as index
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
```

```python
beml_df.head()
```

```python
#import matlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(5,3))
#plt.plot(beml_df.index, beml_df['Close'])
plt.plot(glaxo_df.index, glaxo_df['Close'])
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Close Price of Glaxo')
plt.show()

```

```python
plt.figure(figsize=(5,3))
plt.plot(beml_df.index, beml_df['Close'])
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Close Price of BEML')
plt.show()
```

```python
#adding a gain column
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods=1)
beml_df['gain'] = beml_df.Close.pct_change(periods=1)
```

```python
glaxo_df.head()
```

```python
# droping the first row with NaN. inplace will remove the row with nan permernantly
beml_df.dropna(inplace=True )
glaxo_df.dropna(inplace=True )
```

```python
#plot the gains
plt.figure(figsize=(5,3))
plt.plot(glaxo_df.index, glaxo_df['gain'])
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of Glaxo')
plt.show()
```

```python
#plot the gains
plt.figure(figsize=(5,3))
plt.plot(beml_df.index, beml_df['gain'])
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```

```python
#plot the gains
plt.figure(figsize=(5,3))
plt.plot(beml_df.index, beml_df['gain'],c='red')
plt.plot(glaxo_df.index, glaxo_df['gain'],c='green')
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```
