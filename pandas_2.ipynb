```python
#creating a data frame
df1 = pd.DataFrame({
  'key':['k0','k1','k2','k3'],
  'key1':['k0','k1','k2','k3'],
  'Name':['Abhi','Ayushi','Dhiraj','pani'],
    'age':[17,14,12,52]
})

df2 = pd.DataFrame({
    'key':['k0','k1','k2','k3'],
    'key1':['k0','k1','k2','k3'],
    'Address':['Nagpur','Kanpur','Allahabad','Kannuaj'],
    'Qualification':['BTech1','BE1','ME1','MTech1']
})
print(df1, '\n\n', df2)
```

```python
#creating a data frame
df1 = pd.DataFrame({
  'key':['k0','k1','k2','k3'],
  'key1':['k0','k1','k0','k1'],
  'Name':['Abhi','Ayushi','Dhiraj','pani'],
    'age':[17,14,12,52]
})

df2 = pd.DataFrame({
    'key':['k0','k1','k2','k3'],
    'key1':['k0','k0','k0','k0'],
    'Address':['Nagpur','Kanpur','Allahabad','Kannuaj'],
    'Qualification':['BTech1','BE1','ME1','MTech1']
})
print(df1, '\n\n', df2)
```

```python
#creating dataset
df = pd.DataFrame({
    'Name':['Abhi','Ayushi','Dhiraj','Hitesh','Ravi','Pani','Uma'],
    'Ctg':['A', 'A' , 'C', 'B', 'B', 'B','C'],
    'Val':np.random.randn(7).round(2),
    'Val2':np.random.randint(1,10,size=7)
})
print(df)
```

```python
#logical operater -filtering
df[df['Val']>0.5]
```

```python
#logically on string values
df[df['Name'] > 'Ravi']
```

```python
#adding multiple conditions while filtering | - or operator
df[(df['Val']>0.7) | (df['Val2']==1)]
```

```python
#adding multiple conditions while filtering &- and operator
df[(df['Val']>0.7) & (df['Val2']==1)]
```

```python
#isin operator
names = ['Abhi','Hitesh', 'Pani']
df[df['Name'].isin(names)]
```

```python
#str accessor
df[df['Name'].str.startswith('A')]
```

```python
#str accessor
df[df['Name'].str.endswith('i')]
```

```python
#str contains
df[df['Name'].str.contains('y')]
```

```python
#tilde operator negation operator
df[~df['Name'].str.contains('y')]
```

```python
#nlargest
df.nlargest(3,'Val2')
```

```python
#nsmallest()
df.nsmallest(2,'Val')
```

```python
#creating the dataframe
df = pd.DataFrame({
    'Product':['Carrot','Braccoli','Banana','Banana','Beans','Orange','Braccoly','Beans'],
    'Category':['Vegetable','Vegetable','Fruit','Fruit','Vegetable','Fruit','Vegetable','Vegetable'],
    'Quantity':[8,5,3,4,5,9,11,8],
    'Amount':[270,200,400,500,600,400,190,200]
})

```

```python
#pivot
pivot = df.pivot_table(index='Category',values='Amount',aggfunc='sum')
print(pivot)
```

```python
#pivot
pivot = df.pivot_table(index=['Product','Category'],values=['Amount','Quantity'],aggfunc=['sum','median'])
print(pivot)
```
