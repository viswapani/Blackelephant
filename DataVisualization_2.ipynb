```python
#Catplot --> catagorical scatterplot
#Getting the data
excercise = sns.load_dataset('exercise', index_col=0)
excercise.head()
```

```python
#catplot (continued ...)
sns.catplot(x='time',y='pulse',data=excercise,hue='kind')
# hue is seperating values by color
```

```python
# distplot() - used to get density curve along with histogram
plt.figure(figsize=(5,3))
sns.distplot(excercise['pulse'],kde=False,color='red')
#kernal desity estimate
#probability density function
```

```python
sns.distplot(excercise['pulse'],kde=True,color='green')
```

```python
#loding tips dataset
df = sns.load_dataset('tips')
df.head()
```

```python
#pair plot used to get bird's eye view of the dataset
sns.pairplot(df,hue='smoker',palette='coolwarm')
```

```python
#joint plot used for analyzing two variables
#along with analysizing single variable as well
plt.figure(figsize=(9,3))
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
plt.figure(figsize=(9,3))
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
plt.figure(figsize=(9,3))
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
plt.figure(figsize=(9,3))
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```
